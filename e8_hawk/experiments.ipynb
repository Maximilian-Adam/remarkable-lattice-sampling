## Basic Signing and Verification

```python
message = "E8 signing-stage demo"
salt = bytes.fromhex("00112233445566778899aabbccddeeff")

public_material = make_toy_e8_public_material(k=1)
secret_material = make_toy_e8_secret_material(public_material)

sig_full = sign(message, secret_material, salt=salt)
sig = compact_signature(sig_full)
details = verify(message, sig, public_material, return_details=True)

print("valid:", details["valid"])
print("w norm squared:", details["w_norm_sq"])
print("bound:", details["norm_bound_sq"])
print("sym-break:", details["checks"]["symbreak"])
print("attempts:", sig_full["attempts"])
print("relation w = h - 2s:", sig_full["relation_h_minus_2s"])
```

## Sign Symmetry Demonstration

```python
bad = dict(sig)
bad["s"] = sig_full["h"] - sig_full["s"]
bad_details = verify(message, bad, public_material, return_details=True)

print("transformed valid:", bad_details["valid"])
print("sym-break passes:", bad_details["checks"]["symbreak"])
```
